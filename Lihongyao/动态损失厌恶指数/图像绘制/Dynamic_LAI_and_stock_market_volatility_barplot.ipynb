```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pylab
import matplotlib.ticker as mtick
plt.rcParams['axes.unicode_minus'] = False 
```

```python
data = pd.read_excel('滑动60天分组柱形图数据.xlsx')
data.head()
```

```python
from matplotlib import cm
fig = plt.figure(figsize=(4,4), dpi=300)

x_1 = data['区间股价涨跌幅均值']
x_2 = data['区间股价涨跌幅标准差']
x_3 = data['区间涨跌幅绝对值平均值的平均值']
y_data = ['1','2','3']
colors = cm.Reds([0.3, 0.5, 0.7])
cm.Reds([0, 0.1, 0.8])
```

```python
import seaborn as sns

sns.set_style("white")


plt.figure(figsize=(4, 4))
# plt.bar(data["LAI"], x_1, width=0.5, color = colors)
# plt.plot(data["LAI"], x_1,color='#71020e',marker='o', linestyle='dashed', linewidth=2, markersize=9)
plt.bar(["I", "II", "III"], x_1, width=0.5, color = colors)
plt.plot(["I", "II", "III"], x_1,color='#71020e',marker='o', linestyle='dashed', linewidth=2, markersize=9)

plt.ylabel('$RSW_{mean}$', fontsize = 15)
plt.xlabel('Levels of $DLAI$',fontsize = 15)

plt.yticks(fontproperties = 'Times New Roman',size = 14)
plt.xticks(fontproperties = 'Times New Roman',size = 14)
#plt.savefig('a_bar.png',bbox_inches="tight",dpi=300)
plt.tight_layout()
# plt.show()
plt.savefig("RSW1.pdf", dpi=300)
```

```python
sns.set_style("white")

plt.figure(figsize=(4, 4))
plt.bar(["I", "II", "III"], x_2, width=0.5, color = colors)
plt.plot(["I", "II", "III"], x_2,color='#71020e',marker='o', linestyle='dashed', linewidth=2, markersize=9)

plt.ylim(1.5, 2.1)
plt.ylabel('$RSW_{sd}$', fontsize = 15)
plt.xlabel('Levels of $DLAI$',fontsize = 15)
plt.yticks(fontproperties = 'Times New Roman',size = 14)
plt.xticks(fontproperties = 'Times New Roman',size = 14)
#plt.savefig('b_bar.png',bbox_inches="tight",dpi=300)
plt.tight_layout()
# plt.show()
plt.savefig("RSW2.pdf", dpi=300)
```

```python
sns.set_style("white")

plt.figure(figsize=(4, 4))
plt.bar(["I", "II", "III"], x_3, width=0.5, color = colors)
plt.plot(["I", "II", "III"], x_3,color='#71020e',marker='o', linestyle='dashed', linewidth=2, markersize=9)

ax = plt.gca()
ax.set_ylim(1.2, 1.5)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))
plt.ylabel('$|RSW|_{mean}$', fontsize = 15)
plt.xlabel('Levels of $DLAI$',fontsize = 15)
plt.yticks(fontproperties = 'Times New Roman',size = 14)
plt.xticks(fontproperties = 'Times New Roman',size = 14)
#plt.savefig('c_bar.png',bbox_inches="tight",dpi=300)
plt.tight_layout()
# plt.show()
plt.savefig("RSW3.pdf", dpi=300)
```
