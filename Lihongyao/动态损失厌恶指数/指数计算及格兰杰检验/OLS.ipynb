```python
df = pd.read_csv("DLAI.csv", parse_dates=["start_date", "end_date"])
data = df[df.LAI>-100]
# data["LAI_lag_1"] = data["LAI"].shift(1)
# data["LAI_lag_4"] = data["LAI"].shift(4)
# data["LAI_lag_8"] = data["LAI"].shift(8)
# data["LAI_lag_12"] = data["LAI"].shift(12)
# data["class_lag_1"] = data["classification"].shift(1)
data
```

```python
import pendulum

def add_day(dt, days):
    return pendulum.instance(dt).add(days=days).to_date_string()
```

```python
sh_data = pd.read_excel("上证指数.xlsx", parse_dates=["dt"])
sh_data["r"] = (sh_data["index"] - sh_data["index"].shift(1)) / sh_data["index"].shift(1) * 100
# sh_data["r"] = np.log(sh_data["index"]) - np.log(sh_data["index"].shift(1))

new_data = []

for i, row in data.iterrows():
    row_df = sh_data[(sh_data.dt >= row.start_date) & (sh_data.dt < row.end_date)]
    # print(row_df["r"].mean(), row_df["r"].std(), row_df["r"].abs().mean())
    row_df_5 = sh_data[(sh_data.dt >= add_day(row.start_date,  5)) & (sh_data.dt <= add_day(row.end_date, 5))]
    row_df_10 = sh_data[(sh_data.dt >= add_day(row.start_date,  10)) & (sh_data.dt <= add_day(row.end_date, 10))]
    row_df_20 = sh_data[(sh_data.dt >= add_day(row.start_date,  20)) & (sh_data.dt <= add_day(row.end_date, 20))]
    row_df_30 = sh_data[(sh_data.dt >= add_day(row.start_date,  40)) & (sh_data.dt <= add_day(row.end_date, 30))]
    row_df_40 = sh_data[(sh_data.dt >= add_day(row.start_date,  40)) & (sh_data.dt <= add_day(row.end_date, 40))]
    row_df_50 = sh_data[(sh_data.dt >= add_day(row.start_date,  40)) & (sh_data.dt <= add_day(row.end_date, 50))]
    row_df_60 = sh_data[(sh_data.dt >= add_day(row.start_date,  60)) & (sh_data.dt <= add_day(row.end_date, 60))]
    # row_df = sh_data[(sh_data.dt >= row.start_date) & (sh_data.dt < row.end_date)]

    new_data.append({
        "start_date": row["start_date"],
        "end_date": row["end_date"],
        "LAI": row["LAI"],
        "class": row["classification"],
        "RSW_1": row_df["r"].mean(),
        "RSW_2": row_df["r"].std(),
        "RSW_3": row_df["r"].abs().mean(),
        "RSW_1_5": row_df_5["r"].mean(),
        "RSW_2_5": row_df_5["r"].std(),
        "RSW_3_5": row_df_5["r"].abs().mean(),
        "RSW_1_10": row_df_10["r"].mean(),
        "RSW_2_10": row_df_10["r"].std(),
        "RSW_3_10": row_df_10["r"].abs().mean(),
        "RSW_1_20": row_df_20["r"].mean(),
        "RSW_2_20": row_df_20["r"].std(),
        "RSW_3_20": row_df_20["r"].abs().mean(),
        "RSW_1_30": row_df_30["r"].mean(),
        "RSW_2_30": row_df_30["r"].std(),
        "RSW_3_30": row_df_30["r"].abs().mean(),
        "RSW_1_40": row_df_40["r"].mean(),
        "RSW_2_40": row_df_40["r"].std(),
        "RSW_3_40": row_df_40["r"].abs().mean(),
        "RSW_1_50": row_df_50["r"].mean(),
        "RSW_2_50": row_df_50["r"].std(),
        "RSW_3_50": row_df_50["r"].abs().mean(),
        "RSW_1_60": row_df_60["r"].mean(),
        "RSW_2_60": row_df_60["r"].std(),
        "RSW_3_60": row_df_60["r"].abs().mean(),
    })
    
new_df = pd.DataFrame(new_data, index=None).dropna()
# new_df = new_df[new_df.LAI_lag_1 >= -3]
new_df

# new_df = data.apply(lambda row: {
#     "start_date": row["start_date"],
#     "end_date": row["end_date"],
#     "LAI_lag_1": row["LAI_lag_1"],
#     "RSW_1": sh_data[(sh_data.dt >= row.start_date) & (sh_data.dt <= row.end_date)]["r"].mean(),
#     "RSW_2": sh_data[(sh_data.dt >= row.start_date) & (sh_data.dt <= row.end_date)]["r"].std(),
#     "RSW_3": sh_data[(sh_data.dt >= row.start_date) & (sh_data.dt <= row.end_date)]["r"].abs().mean()
# }, axis=1)

# new_df = new_df.iloc[:-1, ]
# new_df
```
